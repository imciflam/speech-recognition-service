```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
log_csvs = sorted(os.listdir('/content/gdrive/My Drive/logs'))
import matplotlib.gridspec as gridspec
gs1 = gridspec.GridSpec(26,5)
gs1.update(wspace=0.025, hspace=0.05)
```

```python
labels = ['Conv 1D', '', 'Conv 2D', '',  'CRNN','', 'LSTM','']
colors = ['b', 'g' , 'r' , 'c' , 'm' , 'y' , 'k' , '#a88fc9' ]
```

```python
fig, ax = plt.subplots(1, 4, sharey=True, figsize=(20,5), dpi = 200) 
myEnum = enumerate(zip(log_csvs, labels, colors))  
 
i = 0
while i <= 4:
  first_tuple = next(myEnum)
  second_tuple = next(myEnum) 
  csv_path = os.path.join('..', '/content/gdrive/My Drive/logs', first_tuple[1][0])
  csv_path2 = os.path.join('..', '/content/gdrive/My Drive/logs', second_tuple[1][0])
  df = pd.read_csv(csv_path)
  df2 = pd.read_csv(csv_path2) 
  ax[i].set_title(first_tuple[1][1], size=16)
  ax[i].plot(df.accuracy, color=first_tuple[1][2], label='train')
  ax[i].plot(df.val_accuracy, ls='--', color=first_tuple[1][2], label='test')
  ax[i].plot(df2.accuracy, color=second_tuple[1][2], label='noisy_train')
  ax[i].plot(df2.val_accuracy, ls='--', color=second_tuple[1][2], label='noisy_test')
  ax[i].legend(loc='upper left')
  ax[i].tick_params(axis='both', which='major', labelsize=12)
  ax[i].set_ylim([0,1.0])     
  i += 1
 
fig.text(0.5, 0.02, 'Epochs', ha='center', size=14)
fig.text(0.08, 0.5, 'Accuracy', va='center', rotation='vertical', size=14) 
fig.tight_layout()
plt.show()
```
